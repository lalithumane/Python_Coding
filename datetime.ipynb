# we have lot of functions in datetime module but we are learning datetime function from datetime module 

```python
from datetime import datetime    # by this way way we can directly call function from module 
```

```python
now = datetime.now()         # datetime.now()   function gives exact current date and time 
print(now)      
```

```python
day = now.day               # now.day   function gives current date by extracting it from current datetime 
print(day)

month = now.month               # now.month   function gives current month by extracting it from current datetime 
print(month)

year = now.year               # now.year   function gives current year by extracting it from current datetime 
print(year)

hour = now.hour               # now.hour   function gives current hour by extracting it from current datetime 
print(hour)

minute = now.minute               # now.minute   function gives current minute by extracting it from current datetime 
print(minute)

second = now.second               # now.second   function gives current second by extracting it from current datetime 
print(second)
```

### Timestamp or Unix timestamp is the number of seconds elapsed from 1st of January 1970 UTC.

```python
timestamp = now.timestamp()                 # Timestamp or Unix timestamp is the number of seconds elapsed from 1st of January 1970 UTC.
print(timestamp)                  
```

# String to Time Using strptime

```python
from datetime import datetime
date_string = "6 December, 1992"           # mark where is comma and format of string because we have to feed strptime characters in that same manner  
print("Date_String :--> ",date_string)
date_object = datetime.strptime(date_string, "%d %B, %Y" )      #character feeded in same format as above mark comma also
print(date_object)
```

# Using date from datetime

```python
from datetime import date
d = date(2022, 7, 10)
print(d)
print('Current date:', d.today())    # 2019-12-05
# date object of today's date
today = date.today()
print("Current year:", today.year)   # 2019
print("Current month:", today.month) # 12
print("Current day:", today.day)     # 5
```

# Time Objects to Represent Time

```python
from datetime import time
# time(hour = 0, minute = 0, second = 0)
a = time()
print("a =", a)
# time(hour, minute and second)
b = time(5, 52, 50)
print("b =", b)
# time(hour, minute and second)
c = time(hour=10, minute=30, second=50)
print("c =", c)
# time(hour, minute, second, microsecond)
d = time(10, 30, 50, 200555)
print("d =", d)
```

# Difference Between Two Points in Time Using

```python
today = date(year=2022, month=10, day=7)
new_year = date(year=2023, month=1, day=1)
time_left_for_newyear = new_year - today
# Time left for new year:  27 days, 0:00:00
print('Time left for new year: ', time_left_for_newyear)

t1 = datetime(year = 2022, month = 10, day = 7, hour = 0, minute = 59, second = 0)
t2 = datetime(year = 2023, month = 1, day = 1, hour = 0, minute = 0, second = 0)
diff = t2 - t1
print('Time left for new year:', diff) # Time left for new year: 26 days, 23: 01: 00
```

 # Difference Between Two Points in Time Using timedelata

```python
from datetime import timedelta
t1 = timedelta(weeks=12, days=10, hours=4, seconds=20)
t2 = timedelta(days=7, hours=5, minutes=3, seconds=30)
t3 = t1 - t2
print("t3 =", t3)
```
