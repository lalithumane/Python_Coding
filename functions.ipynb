```python
# Declare a function named sum_of_odds. 
# It takes a number parameter and it adds all the odd numbers in that range.

def sum_of_odds(n):
    odd = 0
    for i in range(n+1):
        if i%2 !=0:
            odd = odd + i
    return odd

inpu = int(input('Enter range number here :'))
sum_of_odds(inpu)
```

```python
# Declare a function named evens_and_odds . 
# It takes a positive integer as parameter and it counts number of evens and odds in the number.

def evens_and_odds(num):
    # num = int(input('Eneter a number here to find number of Evens and Odds comes in it :'))
    even = 0
    odd = 0
    for i in range(num+1):
        if i %2 ==0:
            even += i
        else:
            odd +=i       
    print("number of even present in given number {} is {}".format(num,even))
    print('Number of odds present in given number {} is {}'.format(num,odd)) 
evens_and_odds(10)
```
